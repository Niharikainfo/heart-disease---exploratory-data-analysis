```python
import os
import numpy as np 
import pandas as pd 
import seaborn as sb
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
df=pd.read_csv('/kaggle/input/heart-failure-prediction/heart.csv')
df.head()
```

## Distribution for numeric columns

```python
sb.displot(df['Age'],kde= True)
```

```python
sb.distplot(df['Age'],kde= True, hist= True, rug = True)
```

```python
sb.distplot(df['Age'],kde= True, hist= True, rug = True, bins= 10)
```

```python
sb.distplot(df['Age'],kde= True, hist= True, rug = True, bins= 5, color = 'red')
```

```python
sb.displot(df['RestingBP'],kde= True, color = 'purple')
```

```python
sb.distplot(df['RestingBP'],kde= True, color = 'purple')
```

```python
sb.displot(df['Cholesterol'],kde= True, color = 'blue')
```

```python
sb.displot(df['MaxHR'],kde= True, color = 'blue')
```
